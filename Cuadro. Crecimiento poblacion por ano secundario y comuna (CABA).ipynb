```python
import pandas as pd
```

```python
pr07 = pd.read_csv('./2007_matricula_priv_com_CABA.csv', index_col=0)
pu07 = pd.read_csv('./2007_matricula_pub_com_CABA.csv', index_col=0)
pr16 = pd.read_csv('./2016_matricula_priv_com_CABA.csv', index_col=0)
pu16 = pd.read_csv('./2016_matricula_pub_com_CABA.csv', index_col=0)
```

```python
import seaborn as sns

cm = sns.light_palette("green", as_cmap=True)

for i, df in enumerate([pr07, pr16, pu07, pu16]):
    print(['Privada 2007', 'Privada 2016', 'Publica 2007', 'Publica 2016'][i])
    show = df.pct_change(axis = 1).round(2)
    display(show.style.bar(align='mid', color=['#d65f5f', '#5fba7d'], vmin = -.3))
    # show.style.background_gradient(cmap=cm)
```
