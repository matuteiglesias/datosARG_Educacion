```python
import pandas as pd
pd.options.display.max_columns = 99
import numpy as np
```

```python
df_list = []
for regimen in ['1', '2']:

    df1 = pd.read_csv('./anuario_2013/clean/cuadro_2.'+regimen+'.10.csv', header = [0, 1, 2], index_col=0, thousands='.')
    df2 = pd.read_csv('./anuario_2013/clean/cuadro_2.'+regimen+'.11.csv', header = [0, 1, 2], index_col=0, thousands='.')
    df3 = pd.read_csv('./anuario_2013/clean/cuadro_2.'+regimen+'.12.csv', header = [0, 1, 2], index_col=0, thousands='.')
    df4 = pd.read_csv('./anuario_2013/clean/cuadro_2.'+regimen+'.13.csv', header = [0, 1, 2], index_col=0, thousands='.')
    df5 = pd.read_csv('./anuario_2013/clean/cuadro_2.'+regimen+'.14.csv', header = [0, 1, 2], index_col=0, thousands='.')

    # Concatenarlos (ubicarlos uno tras otro horizontalmente)
    df = pd.concat([df1, df2, df3, df5], axis = 1, sort = False)

    # Remover filas de totales
    df = df.loc[[i for i in df.index if 'Total' not in i]]
    df_list += [df]
    
df_ = pd.concat(df_list)
ests_disc = df_['Disciplina']
```

```python
### Carreras elegidas en CABA vs GBA-La Plata vs interior Buenos Aires
inst = pd.read_csv('./instituciones.csv', encoding = 'latin')

inst_CABA = inst.loc[inst['Provincia'] == 'CABA']
inst_GBA = inst.loc[(inst['Provincia'] == 'Buenos Aires') & (inst['Región'] == 'Metropolitana')]
inst_PBA = inst.loc[inst['Región'] == 'Bonaerense']
```

```python
# pd.concat([inst_GBA.set_index('Institución')[['Régimen']], ests_disc], axis = 1)
```

```python
df = pd.concat([ests_disc.loc[ests_disc.index.isin(inst_CABA['Institución'])].sum(),
          ests_disc.loc[ests_disc.index.isin(inst_GBA['Institución'])].sum(),
          ests_disc.loc[ests_disc.index.isin(inst_PBA['Institución'])].sum()], axis = 1)
df.columns = ['CABA', 'GBA', 'PBA']
ests = df.loc[df.index.get_level_values(1) == 'EST']
```

```python
from numpy import round
round(100*ests.sort_values(by = 'CABA', ascending = False)/ests.sum(), 2)
```
