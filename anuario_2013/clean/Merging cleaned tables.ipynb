### Regimen (estatal/privado) de las instituciones

```python

df_ = pd.read_csv('./cuadro_1.1.20.csv', header = [0, 1], index_col=0, thousands='.')
dfA = pd.read_csv('./cuadro_1.2.1.a.csv', header = [0, 1], index_col=0)
dfB = pd.read_csv('./cuadro_1.2.1.b.csv', header = [0, 1], index_col=0)

dfA = dfA.T.loc[dfA.columns.get_level_values(1) == 'Abs.'].T#.loc[]
dfA.columns = dfA.columns.droplevel(1)
dfA = dfA.loc[[i for i in dfA.index if 'Total' not in i]]

dfB = dfB.T.loc[dfB.columns.get_level_values(1) == 'Abs.'].T#.loc[]
dfB.columns = dfB.columns.droplevel(1)
dfB = dfB.loc[[i for i in dfB.index if 'Total' not in i]]
```

#### De paso obtenemos instituciones por ambito

```python
institucionesA = pd.DataFrame(dfA.index, columns= ['Institución'])
institucionesA['regimen'] = 'Estatal'

institucionesB = pd.DataFrame(dfB.index, columns= ['Institución'])
institucionesB['regimen'] = 'Privado'

instituciones = pd.concat([institucionesA, institucionesB]).sort_values(by = 'Institución').reset_index(drop = True)
instituciones.to_csv('./instituciones_noreg.csv', index = False)

instituciones.head()
```

## Evolucion anual

```python
# Estos cuadros tienen la evol. temporal pero cuenta por separado ingresos, egresos y estudiantes.
# Vamos a juntarlos

df_est = pd.read_csv('./cuadro_2.1.1.csv', header = [0, 1], index_col=0, thousands='.')
df_est_p = pd.read_csv('./cuadro_2.2.1.csv', header = [0, 1], index_col=0, thousands='.')
df_est = pd.concat([df_est, df_est_p])
df_ni = pd.read_csv('./cuadro_2.1.2.csv', header = [0, 1], index_col=0, thousands='.')
df_ni_p = pd.read_csv('./cuadro_2.2.2.csv', header = [0, 1], index_col=0, thousands='.')
df_ni = pd.concat([df_ni, df_ni_p])
df_egr = pd.read_csv('./cuadro_2.1.3.csv', header = [0, 1], index_col=0, thousands='.')
df_egr_p = pd.read_csv('./cuadro_2.2.3.csv', header = [0, 1], index_col=0, thousands='.')
df_egr = pd.concat([df_egr, df_egr_p])

df = pd.DataFrame(pd.concat([df_est, df_ni, df_egr], axis = 1, sort = False).stack()).reset_index()
df.columns = ['Institución','Año','Egresados','Estudiantes','Nuevos Inscriptos']
# df = df.loc[[i for i in df.index if 'Total' not in i]] # Remover rows de totales
df = df.loc[~df['Institución'].str.lower().str.contains('total')].reset_index(drop = True)
```

## Numeros segun censo:

```python
educ_censo = pd.read_csv('./../../../Datos_censo/Preguntas/niveles_educativos.csv')
radio_ref = pd.read_csv('./../../../Datos_censo/Radios_censales/radio_ref.csv')
radio_ref = radio_ref.loc[radio_ref.AGLOMERADO.isin([2, 32, 33])]
```

```python
# Univs sin Terminar por partido
merged = educ_censo.merge(radio_ref, on = 'RADIO_REF_ID')
univs_sinT = merged.groupby(['NOMPROV', 'NOMDPTO', 'DPTO']).sum()['72 Universitario sin T'].sort_values()
univs_sinT
```

```python
# Agregar parte espacial
import geopandas as gpd
rad_geom = gpd.read_file('./../../../Datos_censo/Radios_censales/radios_w_geometry.shp')
radio_ref['LINK'] = radio_ref['radio'].astype(str).str.zfill(9)

rad_geom = gpd.GeoDataFrame(radio_ref.merge(rad_geom[['LINK', 'geometry']]))

frac_geom = rad_geom.dissolve(by = ['NOMPROV', 'NOMDPTO', 'IDFRAC'])[['geometry']]
```

```python
ests_by_frac = merged.groupby(['NOMPROV', 'NOMDPTO', 'IDFRAC'])[['72 Universitario sin T']].sum()
```
